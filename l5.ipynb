```python
import matplotlib.pyplot as plt

def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return comparisons
    return comparisons

list_sizes = range(1, 1001)
comparisons = []

for size in list_sizes:
    arr = list(range(size))
    target = size
    comparisons.append(linear_search(arr, target))

plt.plot(list_sizes, comparisons)
plt.title("Linear Search: Number of Comparisons vs. List Size")
plt.xlabel("List Size")
plt.ylabel("Number of Comparisons")
plt.grid(True)
plt.show()
```
